```python
import numpy as np
import matplotlib.pyplot as plt

# Q2

x = np.linspace(0, 1, 50)
y_log = np.log(x + 1e-10)
y_exp = np.exp(x)

plt.scatter(x, y_log, label='Natural Log')
plt.scatter(x, y_exp, label='Exponential')
plt.xlabel('x values')
plt.ylabel('Function values')
plt.title('Natural Log and Exponential Functions')
plt.legend(loc='lower right')
plt.show()

x = np.linspace(-6.5, 6.5, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='Sine Function')
plt.plot(x, y_cos, label='Cosine Function')
plt.xlabel('x values')
plt.ylabel('Function values')
plt.title('Sine and Cosine Functions')
plt.legend(loc='lower left')
plt.show()
```
